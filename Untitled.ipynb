```python
null_columns = sharks.isnull().mean()*100
null_columns = null_columns[null_columns > 99]
null_columns
```

```python
cols_eliminate = list(null_columns.index)
cols_eliminate
```

```python
sharks_clean= sharks.drop(cols_eliminate, axis=1)
sharks_clean
```
